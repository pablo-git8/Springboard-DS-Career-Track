**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column.

```python
#Adding Year column extracting year from Month column at the DF
properties.loc[:,'Year'] = pd.DatetimeIndex(properties['Month']).year 
```

```python
properties.head() #check the result on adding the Year column
```

```python
#Getting a df for average mean prices in one year per each Borough
years = [1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
l = []

#Loop over the years
for i in years:
    mean_year = properties.groupby(['Borough', 'Year']).apply(lambda x: x.loc[x.Year == i])
    #Loop over the official Boroughs
    for j in official_b:
        m_year = mean_year[mean_year['Borough'] == j].agg({'Average Price':'mean'})
        l.append([j, i, m_year['Average Price']]) #Create the list for the DF
```

```python
#Create the DF and groupby Boroughs and Years
dfm = pd.DataFrame(l, columns =['Borough', 'Year', 'Average Price'])
dfm.groupby(['Borough','Year']).mean()
```

```python
sub_prop = dfm[dfm['Borough']=='Brent'] #Subsetting Brent Borough
```

```python
#Making a basic line plot of the subseted DF
plt.figure(figsize=(8,8))
plt.plot('Year','Average Price',data = sub_prop)
plt.title('Average Price from 1995 to 2020 in Brent')
plt.xlabel('Years')
plt.ylabel('Average Price')
plt.show()
#Since there are 12 values for each year, we se the prices in a plot like this in vertical line
#If we take the avergae price all over the year we can get a better plot
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

```python
properties['Year'] = pd.DatetimeIndex(properties['Month']).year #Adding the year column to the properties DF
df = properties.copy() #Copy DF for the next calculations
```

```python
prices = {}
for i in official_b:
    #Average averge price for a property in year 1998
    avp_1998 = df[(df['Borough'] == i) & (df['Year'] == 1998)].agg({'Average Price':'mean'}) 
    #Average averge price for a property in year 1998
    avp_2018 = df[(df['Borough'] == i) & (df['Year'] == 2018)].agg({'Average Price':'mean'}) 
    #Compute the ratio in that borough
    ratio = avp_2018/avp_1998 
    #Create the dictionary to store the ratios
    prices[i] = ratio

pdf = pd.DataFrame(prices).transpose().reset_index().rename(columns={'index':'Borough', 'Average Price':'Increment Ratio'})\
.sort_values(by='Increment Ratio', ascending=False)
```

```python
pdf #check Prices DF
```

```python
#Making a basic bar plot of the Prices DF
plt.figure(figsize=(12,5))
plt.bar('Borough','Increment Ratio',data = pdf)
plt.title("Borough's Increment Ratio from 1998 to 2018")
plt.xlabel('Borough')
plt.ylabel('Increment Ratio')
plt.xticks(rotation=80)
plt.show()
```
